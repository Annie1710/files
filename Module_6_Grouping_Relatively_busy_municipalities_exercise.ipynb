```python
provincies = "https://raw.githubusercontent.com/WincAcademy/practice_data/main/data/geo/provinces.csv"
provincies = pd.read_csv(provincies)	
provincies.rename(columns = {'inhabitants':'inhabitants_province'}, inplace = True)
# provincies
```

```python
totaal = pd.merge(nederland, provincies, on="province")
totaal
```
